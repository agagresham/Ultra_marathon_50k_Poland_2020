```python
# seasons for the data slower in summer than winter?
# spring 3 - 5
# summer 6-8
# autumn 9 - 11
# winter 12 - 2
# split between two decimals
df_poland['race_month'] = df_poland['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df_poland.head()
```

```python
df_poland['race_season'] = df_poland['race_month'].apply(lambda x: 'Winter'if x>11 else "Autumn" if x > 8 else 'Summer' if x >5 else "Spring" if x > 2 else "Winter")
```
